```python
import matplotlib.pyplot as plt
import numpy as np

dir="../../build-tdvp-Desktop-Release/example/"
L=20
for orb in ["real_space","star"]:
    dat=np.loadtxt(f"{dir}/irlm_{orb}_L{L}.txt",skiprows=1)
    plt.plot(dat[2:,0],np.abs(dat[2:,2]-dat[1,2])+1e-16, label=orb)
plt.yscale("log")
plt.xlabel("time")
plt.ylabel("energy")
plt.legend();
```

```python
plt.xlabel("time")
plt.ylabel("mps bond dimension")
plt.title(f"IRLM L={L}")
for orb in ["real_space","star"]:
    dat=np.loadtxt(f"{dir}/irlm_{orb}_L{L}.txt",skiprows=1)
    plt.plot(dat[:,0],dat[:,1],label=orb)    
plt.legend();
```

```python
import matplotlib.pyplot as plt
import numpy as np

L=100
n0=np.loadtxt(f"output_star_L{L}_n0.txt")
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.xlabel("time")
plt.ylabel("|n0-0.5|")
#plt.yscale("log")
plt.plot([x-0.1 for x in range(0,len(n0))],[abs(x-0.5) for x in n0], '-', label=f"old L={L}, U=0.5")

for L in [100,200,201]:
    n02=np.loadtxt(f"L{L}/n0.txt")
    plt.plot(n02[:,0],[abs(x-0.5) for x in n02[:,3]], '-', label=f"L={L}, U=0.5")

plt.legend()
```

```python
L=200
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.ylabel("evals <cicj>")
plt.yscale("log")
for t in [40,70]:
    eval=np.loadtxt(f"L{L}/eval_t{t}.txt")
    activ=[min(x,1-x) for x in eval]
    plt.plot(activ, 'o--',label=f"L={L}, U=0.5, t={t}")
plt.xlim(0,110)
plt.legend()
```

```python
L=100
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.ylabel("evals <cicj>")
plt.yscale("log")
for t in [0,49]:
    eval=np.loadtxt(f"output_star_L{L}_eval_t{t}.txt")
    activ=[min(x,1-x) for x in eval]
    plt.plot(activ, 'o--',label=f"L={L}, U=0.5, t={t}")
plt.legend()
```
