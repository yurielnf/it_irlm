```python
import matplotlib.pyplot as plt
import numpy as np

L=100
n0=np.loadtxt(f"output_star_L{L}_n0.txt")
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.xlabel("time")
plt.ylabel("|n0-0.5|")
plt.yscale("log")
#plt.plot([x-0.1 for x in range(0,len(n0))],[abs(x-0.5) for x in n0], '-', label=f"old L={L}, U=0.5")

for L in []:   #100,101,200,202,300,203]:
    n02=np.loadtxt(f"L{L}/n0.txt")
    plt.plot(n02[:,0],[abs(x-0.5) for x in n02[:,3]], '-', label=f"L={L}, U=0.5")

L=30
for U in [0.25]:#[0.5, 0.25, 0.252]:
    n02=np.loadtxt(f"wilson{L}U{U}/n0.txt")
    plt.plot(n02[:,0],[abs(x-0.5) for x in n02[:,3]], '--', label=f"wilson star L={L}, U={U}")

U=0.25
for t in [1, 0.5, 0.75]:
    n02=np.loadtxt(f"wilson_chain{L}U{U}t{t}/n0.txt")
    plt.plot(n02[:,0],[abs(x-0.5) for x in n02[:,3]], '--', label=f"wilson chain L={L}, U={U} t0={t}")

for t in [0.75]:
    n02=np.loadtxt(f"wilson_chain{L}U{U}t{t}star/n0.txt")
    plt.plot(n02[:,0],[abs(x-0.5) for x in n02[:,3]], '--', label=f"wilson chain->star L={L}, U={U} t0={t}")

plt.legend()
```

```python
L=200
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.ylabel("evals <cicj>")
plt.yscale("log")
for t in [40,70]:
    eval=np.loadtxt(f"L{L}/eval_t{t}.txt")
    activ=[min(x,1-x) for x in eval]
    plt.plot(activ, 'o--',label=f"L={L}, U=0.5, t={t}")
plt.xlim(0,110)
plt.legend()
```

```python
L=202
plt.title("IRLM: quench from ed=-10 to ed=0")
plt.ylabel("evals <cicj>")
plt.yscale("log")
for t in [98]:
    for lab in ["eval","ni"]:
        eval=np.loadtxt(f"L{L}/{lab}_t{t}.txt")
        eval=sorted(eval)
        activ=[min(x,1-x) for x in eval]
        plt.plot(activ, 'o--',label=f"{lab} L={L}, U=0.5, t={t}")
plt.legend()
```
